# MinMaxScaling(Normalization) 

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("C:\\Users\\saurabh\\Desktop\\pyth\\Social_Network_Ads.csv")
df.head() 
```

```python
df = df.drop(columns = ['Gender'])
```

```python
x = df.drop(columns = ['Purchased'] , axis=1 ) 
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
mn = MinMaxScaler() 
```

```python
x_train_scaled = mn.fit_transform(x_train) 
x_test_scaled = mn.transform(x_test) 
```

```python
x_train_new = pd.DataFrame(x_train_scaled , columns = x_train.columns) 
```

```python
np.round(x_train.describe(), 1) 
```

```python
np.round(x_train_new.describe() , 1) 
```
