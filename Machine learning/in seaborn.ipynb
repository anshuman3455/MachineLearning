#  Opening a csv file form an URL

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x=data['total_bill'],y=data['tip'])
```

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter')
```

```python
sns.relplot(x='total_bill',y='tip',hue = 'smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(10,300),data=data)
```
