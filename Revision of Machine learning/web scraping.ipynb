```python
name = []
rating = []
price = []
storage = []
length = []
camera = []
battery = []

for i in company:
    
    name.append(i.find(class_ ='_4rR01T').text)
    rating.append(i.find(class_ = '_3LWZlK').text)
    price.append(i.find(class_ = '_30jeq3 _1_WHN1').text)
    storage.append(i.find_all(class_ = 'rgWa7D')[0].text.strip())
    length.append(i.find_all(class_ = 'rgWa7D')[1].text)
    camera.append(i.find_all(class_ = 'rgWa7D')[2].text)
    battery.append(i.find_all(class_='rgWa7D')[3].text)
    
df = pd.DataFrame({'name':name,
                 'rating':rating ,
                   'price':price,
                   'Storage':storage,
                   'Lenght':length,
                   'Camera':camera,
                   'Battery':battery})
```
