```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("C:\\Users\\user\\Documents\\data sets\\titanic.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
X_train.head()
```

```python
number = ['Age','Fare']
num_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer()),
    ('scaler',StandardScaler())
])

categorical = ['Embarked','Sex']
cat_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('ohe',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
perprocessor= ColumnTransformer(
        transformers=[
            ('num',num_transformer,number),
            ('cat',cat_transformer,categorical)
        ]
)
```

```python
clf = Pipeline(steps=[
    ('preprocessor',perprocessor),
    ('classifier',LogisticRegression())
])
```

```python
from sklearn import set_config

set_config(display='diagram')
clf
```

```python
parm_grid ={
    'preprocessor__num__imputer__strategy':['mean','median'],
    'preprocessor__cat__imputer__strategy':['most_frequent','constant'],
    'classifier__C':[0.1,1.0,10,100]
}
grid_search = GridSearchCV(clf,parm_grid,cv=10)
```

```python
grid_search.fit(X_train, y_train)

print(f"Best params:")
print(grid_search.best_params_)
```

```python
print(f"Internal CV score: {grid_search.best_score_:.3f}")
```

```python
import pandas as pd

cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results = cv_results.sort_values("mean_test_score", ascending=False)
cv_results[['param_classifier__C','param_preprocessor__cat__imputer__strategy','param_preprocessor__num__imputer__strategy','mean_test_score']]
```
