```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\user\\Downloads\\Titanic-Dataset.csv",usecols = ['Cabin','Ticket','Parch','Survived'])
```

```python
df['Parch'][1]= 'a'
df['Parch'][3]= 'a'
df['Parch'][5]= 'a'
df['Parch'][8]= 'a'
df['Parch'][145]= 'a'
df['Parch'][100]= 'a'
df['Parch'][890]= 'a'
df['Parch'][887]= 'a'
```

```python
df['Parch'].unique()
```

```python
df['Parch_numerical'] = pd.to_numeric(df["Parch"],errors='coerce',downcast='integer')
```

```python
df['Parch_categorical'] = np.where(df['Parch_numerical'].isnull(),df['Parch'],np.nan)
df.head()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
